## Question 1: What do the following distributions signify about samples drawn from it

### a. Gaussian distribution

#### It's symmetrical with single peak. Data has finite mean and variance. The likelihood of the sample point is maximized at the mean. 

### b. Exponential distribution

#### It's right-skewed with long, flat right tails. Data points are more likely to be sampled near smaller values than larger values. Data is usually associated with a constant failure rate. 

### c. Weibull distribution

#### It's either right-skewed or approximately symmetric. It can be resemabling either normal or exponential distribution. The advantage of Weibull is that it can fit skewed data better.

## Question 3. At a 0.05 significance level, test the following hypothesis:  [Hint: Use a Z-test for testing this hypothesis].

###  The AV has more disengagements (automatic and manual) on cloudy days than clear days. 

```python
#Two Proportion Z-Test
p1 = sum_cloudy_disengagements / (sum_miles * (1 - 0.72))
p2 = sum_clear_disengagements / (sum_miles * (0.72))
p = (sum_cloudy_disengagements + sum_clear_disengagements) / sum_miles
z = ((p1 - p2) - 0) / np.sqrt(p * (1-p) * (1/(sum_miles * (1 - 0.72)) + 1/(sum_miles * (0.72))))
p_val = st.norm.sf(abs(z))

print(z, p_val)
```

###  Based on the result of the hypothesis test, what can you conclude about the impact of weather conditions on AV safety?

####  Since p value is smaller than 0.05, we reject the null and conclude that the AV has more disengagements (automatic and manual) on cloudy days than clear days.

# Task 3 - Using the Naive Bayes Model

## Question 1. Though there are 10 different causes for disengagement, they can be grouped into the following 3 classes  (i) Controller, (ii) Perception System, and (iii) Computer System. The mapping from Disengagement Cause to Class is given in the table below. You will use these 3 classes as the labels in the NB model. Modify your pandas data frame to include a Class column.

```python
disengagements['Class'] = [
    'Controller' if x == 'Incorrect behavior prediction of others' or x == 'Recklessly behaving agent' or x =='Unwanted Driver Discomfort'
    else (
    'Perception System' if x == 'Adverse road surface conditions' or x == 'Emergency Vehicle' or x == 'Position Estimation Failure' or x == 'Incorrect Traffic Light Detection'
    else
    'Computer System'
    )
    for x in disengagements['Cause']
]
disengagements.head()
```

## Question 2. Split the data randomly into training and testing (80-20 split). Use the fields Location, Weather, and TypeOfTrigger as features and use Class as assigned in the previous question as the label.

```python
sub_df = disengagements[['Location', 'Weather', 'TypeOfTrigger', 'Class']]
sub_df.insert(0, 'id', range(len(sub_df)))

random_array_ids =  [i for i in range(len(sub_df))]
np.random.shuffle(random_array_ids)
training_amount = int(0.8*len(sub_df))

training_ids = random_array_ids[:training_amount]
testing_ids = random_array_ids[training_amount:]

training_df = sub_df.loc[sub_df['id'].isin(training_ids)][['Location', 'Weather', 'TypeOfTrigger', 'Class']]
testing_df = sub_df.loc[sub_df['id'].isin(testing_ids)][['Location', 'Weather', 'TypeOfTrigger', 'Class']]

training_df.head()
testing_df.head()
```

## Question 3. Using the training dataset, create a NB model to identify the cause of disengagement based on the features Location, Weather, and TypeOfTrigger. Show the conditional probability tables from the training dataset.

```python
#https://en.wikipedia.org/wiki/Naive_Bayes_classifier
p = {} #prob
c = {} #count

row_count = training_df['Class'].count()
c['clear'] = training_df.loc[training_df['Weather'] == 'clear']['Class'].count() 
c['cloudy'] = training_df.loc[training_df['Weather'] == 'cloudy']['Class'].count() 
c['highway'] = training_df.loc[training_df['Location'] == 'highway']['Class'].count() 
c['urban-street'] = training_df.loc[training_df['Location'] == 'urban-street']['Class'].count()
c['automatic'] = training_df.loc[training_df['TypeOfTrigger'] == 'automatic']['Class'].count() 
c['manual'] = training_df.loc[training_df['TypeOfTrigger'] == 'manual']['Class'].count()
c['Perception System'] = training_df.loc[training_df['Class'] == 'Perception System']['Class'].count()
c['Computer System'] = training_df.loc[training_df['Class'] == 'Computer System']['Class'].count()
c['Controller'] = training_df.loc[training_df['Class'] == 'Controller']['Class'].count()

p['clear'] = c['clear'] / row_count
p['cloudy'] = c['cloudy'] / row_count
p['highway'] = c['highway'] / row_count
p['urban-street'] = c['urban-street'] / row_count
p['automatic'] = c['automatic'] / row_count
p['manual'] = c['manual'] / row_count
p['Perception System'] = c['Perception System'] / row_count
p['Computer System'] = c['Computer System'] / row_count
p['Controller'] = c['Controller'] / row_count

c[('Perception System','cloudy')] = training_df.loc[(training_df['Class'] == 'Perception System') & (training_df['Weather'] == 'cloudy')]['Class'].count() 
c[('Perception System','clear')] = training_df.loc[(training_df['Class'] == 'Perception System') & (training_df['Weather'] == 'clear')]['Class'].count()
c[('Perception System','highway')] = training_df.loc[(training_df['Class'] == 'Perception System') & (training_df['Location'] == 'highway')]['Class'].count()
c[('Perception System','urban-street')] = training_df.loc[(training_df['Class'] == 'Perception System') & (training_df['Location'] == 'urban-street')]['Class'].count()
c[('Perception System','automatic')] = training_df.loc[(training_df['Class'] == 'Perception System') & (training_df['TypeOfTrigger'] == 'automatic')]['Class'].count()
c[('Perception System','manual')] = training_df.loc[(training_df['Class'] == 'Perception System') & (training_df['TypeOfTrigger'] == 'manual')]['Class'].count()

p[('Perception System','cloudy')] = c[('Perception System','cloudy')] / c['Perception System'] 
p[('Perception System','clear')] = 1 - p[('Perception System','cloudy')] #c[('Perception System','clear')] / c['Perception System']
p[('Perception System','highway')] = c[('Perception System','highway')] / c['Perception System'] 
p[('Perception System','urban-street')] = 1 - p[('Perception System','highway')] #c[('Perception System','urban-street')] / c['Perception System']
p[('Perception System','automatic')] = c[('Perception System','automatic')] / c['Perception System'] 
p[('Perception System','manual')] = 1 - p[('Perception System','automatic')] #c[('Perception System','manual')] / c['Perception System'] 

c[('Computer System','cloudy')] = training_df.loc[(training_df['Class'] == 'Computer System') & (training_df['Weather'] == 'cloudy')]['Class'].count() 
c[('Computer System','clear')] = training_df.loc[(training_df['Class'] == 'Computer System') & (training_df['Weather'] == 'clear')]['Class'].count()
c[('Computer System','highway')] = training_df.loc[(training_df['Class'] == 'Computer System') & (training_df['Location'] == 'highway')]['Class'].count()
c[('Computer System','urban-street')] = training_df.loc[(training_df['Class'] == 'Computer System') & (training_df['Location'] == 'urban-street')]['Class'].count()
c[('Computer System','automatic')] = training_df.loc[(training_df['Class'] == 'Computer System') & (training_df['TypeOfTrigger'] == 'automatic')]['Class'].count()
c[('Computer System','manual')] = training_df.loc[(training_df['Class'] == 'Computer System') & (training_df['TypeOfTrigger'] == 'manual')]['Class'].count()

p[('Computer System','cloudy')] = c[('Computer System','cloudy')] / c['Computer System'] 
p[('Computer System','clear')] = 1 - p[('Computer System','cloudy')]  #c[('Computer System','clear')] / c['Computer System']
p[('Computer System','highway')] = c[('Computer System','highway')] / c['Computer System'] 
p[('Computer System','urban-street')] = 1 - p[('Computer System','highway')]#c[('Computer System','urban-street')] / c['Computer System']
p[('Computer System','automatic')] = c[('Computer System','automatic')] / c['Computer System'] 
p[('Computer System','manual')] = 1 - p[('Computer System','automatic')]#c[('Computer System','manual')] / c['Computer System'] 

c[('Controller','cloudy')] = training_df.loc[(training_df['Class'] == 'Controller') & (training_df['Weather'] == 'cloudy')]['Class'].count() 
c[('Controller','clear')] = training_df.loc[(training_df['Class'] == 'Controller') & (training_df['Weather'] == 'clear')]['Class'].count()
c[('Controller','highway')] = training_df.loc[(training_df['Class'] == 'Controller') & (training_df['Location'] == 'highway')]['Class'].count()
c[('Controller','urban-street')] = training_df.loc[(training_df['Class'] == 'Controller') & (training_df['Location'] == 'urban-street')]['Class'].count()
c[('Controller','automatic')] = training_df.loc[(training_df['Class'] == 'Controller') & (training_df['TypeOfTrigger'] == 'automatic')]['Class'].count()
c[('Controller','manual')] = training_df.loc[(training_df['Class'] == 'Controller') & (training_df['TypeOfTrigger'] == 'manual')]['Class'].count()

p[('Controller','cloudy')] = c[('Controller','cloudy')] / c['Controller'] 
p[('Controller','clear')] = 1 - p[('Controller','cloudy')]#c[('Controller','clear')] / c['Controller']
p[('Controller','highway')] = c[('Controller','highway')] / c['Controller'] 
p[('Controller','urban-street')] = 1 - p[('Controller','highway')]#c[('Controller','urban-street')] / c['Controller']
p[('Controller','automatic')] = c[('Controller','automatic')] / c['Controller'] 
p[('Controller','manual')] = 1 - p[('Controller','automatic')] #c[('Controller','manual')] / c['Controller'] 

print(p)

def NB(Weather, Location, TypeOfTrigger):
    cks = ['Perception System', 'Computer System', 'Controller']
    evidence = 0
    for ck in cks:
        evidence += p[ck] * p[(ck, Weather)] * p[(ck, Location)] * p[(ck, TypeOfTrigger)]
    ps = []
    for ck in cks:
        ps.append([ck, p[(ck, Weather)] * p[(ck, Location)] * p[(ck, TypeOfTrigger)] * p[ck] / evidence])
    ps = sorted(ps, key = lambda x : x[1])[::-1]
    return ps[0][0]
```

## Question 4. Using the model to predict the cause of the disengagement for the test dataset. Compute the accuracy achieved by your model.

```python
def classify(row):
    return NB(row['Weather'], row['Location'], row['TypeOfTrigger'])
testing_df['NB Class'] = testing_df.apply(classify, axis=1)
expected = list(testing_df['NB Class'])
actual = list(testing_df['Class'])
count = 0
for i in range(len(expected)):
    if expected[i] != actual[i]:
        count += 1
print("NB classifier's accuracy is:")
print(1 - count / len(expected))
```
